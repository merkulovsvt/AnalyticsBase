```python

import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("..\\Datasets\\terrorist-attacks new.csv")
data.head()
```

```python
data.plot(kind='scatter', x='Year', y='Terrorist attacks')
```

```python
data.describe()
```

```python
zaza = data.groupby('Year')['Terrorist attacks'].sum().to_frame()
zaza.plot()
```

```python
a = data.groupby('Year')['Terrorist attacks'].sum()
new_zaza = pd.DataFrame([{'Year': data['Year'].unique()}])
new_zaza.merge(a, on='Year', how='left')
```

```python
a = data.groupby('Year')['Terrorist attacks'].sum()
a.head()
```

```python
sns.histplot(data[['Year', 'Terrorist attacks']], x="Year", y=kde = True, bins = 30)
```

```python
a = pd.DataFrame({'Year': data['Year'].unique(), 'Sum': data.groupby('Year')['Terrorist attacks'].sum()})
a.reset_index(inplace=True, drop=True)
sns.heatmap(data[data['Code'] == 'AFG'][['Year', 'Terrorist attacks']])
```

```python
a = pd.DataFrame({'Year': data['Year'].unique(), 'Amount of attacks': data.groupby('Year')['Terrorist attacks'].sum()})
a.reset_index(inplace=True, drop=True)

plot = sns.barplot(data=a, x=a['Year'], y=a["Amount of attacks"])
import matplotlib.ticker as ticker

plot.xaxis.set_major_locator(ticker.LinearLocator(10))

```
