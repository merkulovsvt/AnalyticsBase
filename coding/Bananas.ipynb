```python
import pandas as pd
import seaborn as sns

data = pd.read_csv("../Datasets/banana_quality.csv")
data.rename(columns=lambda x: x.replace(' ', '_').lower(), inplace=True)
data.head(10)
data.shape
```

```python
for col_name in list(data.select_dtypes(include='float').columns):
    sns.histplot(data=data, x=col_name, kde=True)
```

```python
sns.scatterplot(data=data, x='sweetness', y='ripeness', hue="quality")
```
