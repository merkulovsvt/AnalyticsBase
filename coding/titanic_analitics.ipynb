```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("..\\Datasets\\titanic.csv")
data.head()
```

```python
survived_passengers = data[data['Survived'] == 1]
dead_passengers = data[data['Survived'] == 0]
```

```python
sns.histplot(data=survived_passengers, x='Age', kde=True, hue='Pclass', palette='Set1')
```

```python
final_data = pd.DataFrame({"Pclass": data['Pclass'].unique()})


count_passengers = data.groupby('Pclass')['Age'].count().rename("Passanger count")
survival_rate = data.groupby('Pclass')['Survived'].mean().rename("Survival Rate")

final_data = final_data.merge(count_passengers, on='Pclass')
final_data = final_data.merge(survival_rate, on='Pclass')

mean_age = data.groupby('Pclass')['Age'].mean().rename("Mean age")
median_age = data.groupby('Pclass')['Age'].median().rename("Median age")
# mode_age = (data.groupby('Pclass')['Age'].mod() / data.groupby('Pclass')['Age'].count())

final_data = final_data.merge(mean_age, on='Pclass')
final_data = final_data.merge(median_age, on='Pclass')

# final_data = final_data.merge(mode_age, on='Pclass')


# pd.DataFrame(da.si
final_data.rename({'Pclass': 'Pclass1'}, inplace=True)
final_data.fillna(axiinplace=)
```
